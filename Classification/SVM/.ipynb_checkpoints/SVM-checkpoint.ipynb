<font size='4'> This model is about finding the decision boundary that best (optimal) seperates the different groups of data points. SVM finds the seperating line or hyper plane by finding the maximum margin (max perpendicular distance b/w hyper plane & decision boundary). The algorithm finds the hyper plane by maximizing the distance between the decision boundaries, denoted by $2\epsilon$.</font>

```python
Image('CAPTURE.png', embed=True)
# The 'support vectors' are so called because they "support" or help the algorithm in finding the decision boundaries. In 
# order to understand how SVM finds the support vectors, let's consider an example. Suppose there are 2 classes apples &
# oranges. The algorithm will find the apples that closely resemble oranges, and oranges that closely resemble apples. These
# data points would usually be really close to each other since they share somewhat common charecterstics and they are the 
# ones that are potential support vector candidates. It works in a similar manner in multi-class data. Once potential
# support vector points are found, the algorithm tries to find the hyper plane with maximum margin b/w those support 
# vectors.

# The decision boundaries are  equidistant from the hypyer line/plane (epsilon).
```

<font size='3'>Since the data cannot be seperated by a simple line or linear hyper plane (incase data was 3D or higher), it uses something called a *kernel*. Kernel is a complex mathematical function that transforms data from lower dimension to higher dimension (for eg 2D to 3D or 3D to 4D) in order to be able to classify them using a linear hyper plane. Because usually,non linear data transforms into linearly seperable data when transformed from lower dimension to higher dimension by passing it through the kernel function.</font>

<font size='3' >SVMs (linear or otherwise) inherently do binary classification. However, there are various procedures for extending them to multiclass problems. The most common methods involve transforming the problem into a set of binary classification problems, by one of two strategies:

1. One vs. the rest. For k classes, k binary classifiers are trained. Each determines whether an example belongs to its 'own' class versus any other class. The classifier with the largest output is taken to be the class of the example.
    
    
2. One vs. one. A binary classifier is trained for each pair of classes. A voting procedure is used to combine the outputs.

More sophisticated methods exist.</font>

The **maximum margin classifer** (a classifier that uses maximum margins to create the decision boundaries as discussed above) is **super sensitive to outliers in training data.** To find a hyper plane that is not so sensitive to outliers, we must allow for the model to make misclassifications. Choosing a hyper plane that allows for certain misclassification is an example of Bias/Variance tradeoff.

When we allow for misclassifications, the distance b/w the observations (or decision boundary) and the hyper plane is called a **Soft margin**

The question is, which margin (or how much misclassification to allow) will result in the best Soft Margin or the best hyper plane. We use cross - validation to determine how many misclassifications and observations to allow inside of the Soft Margin to get best classification.

Soft Margin classifier is aka Support Vector Classifiers. :)

**NOTE** Support Vector Classifiers are different from SVM. Support vector classifier can produce only a linear hyper plane. Whereas SVMs is a method that uses functions like *kernel* to trasnform data into higher dimension and then make use of Support Vector Classifiers.

The kernel being used, finds the relationship between each pair of observation in **X** dimension (where X is chosen by the user), and uses that info to find the SVC.

**NOT** The kernel function actually uses a set of slightly complex mathematical calculations that allows it to calculate the relationships b/w each pair of observations as **if** they are in **X** dimensions ; they don't actually transform the data to higher dimensions. This trick of calculating the high dimensional relationships without actually tranforming the data to higher dimensions, is called the **kernel trick**. Different types of kernels include polynomial kernel, RBF kernel etc. The kernel trick reduces the amount of computations and significantly decreases the time to find the SVC.

Once the high dimensional relationships for each point is found, how this information is used in order to find the SVC, is still beyond me :(

<font size='3'> Polynomial kernel : $$ relationship = (x_1 x_2 + r)^d $$ 
</font>

$x_1, x_2$ = refers to datapoints

$r$ = polynomials coefficient

$d$ = degree of polynomial

We want the best(optimal) value for r & d, and therefore we use Cross validation to find that.

Since RBF kernel finds the SVC in infinite dimensions, it's impossible to visualize. However, when using it to predict new observations, it behaves like a weighted Nearest Neighbor model.

RBF kernel determines how much influence each observation in the training data have on the new observation using : 

<font size='4'> $$ relationship = e^{- \gamma(x_1-x_2)^2} $$ </font>

The amount of influence one observation has on another, is a function of squared distance between them.
$\gamma$, which is determined by cross validation, scales the squared distance, thus scaling the influence.

RBF kernel uses the Taylor Series and can be thought of as a constitute of multiple polynomial kernel with r=0 and d increasing form 0 to $\infty$, thus giving you points in infinite dimensions. And then finds the relationship between these 'infinite' dimensional points and uses that info to find the SVC.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
import utils
from sklearn.metrics import accuracy_score, confusion_matrix , precision_recall_curve
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state=1, train_size=0.7)
```

```python
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test)
```

```python
model = SVC(kernel='linear', random_state=1)
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, Y_test)
```

```python
cm = confusion_matrix(Y_test, y_pred)
print(cm)
print("accuracy score :", accuracy_score(Y_test, y_pred))
```

```python
cm_metrics = utils.cm_ratios(cm)

print('accuracy :', cm_metrics.accuracy)
print('misclassification_rate :', cm_metrics.misclassification_rate)
print('TPrate:', cm_metrics.TPrate, '(recall)')
print('TNrate:', cm_metrics.TNrate)
print('precision:', cm_metrics.precision)
print('FPrate:', cm_metrics.FPrate)
print('FNrate:', cm_metrics.FNrate)
```
