#### Multiple Linear Regression (MLR) is just an extension of Simple Linear Regression. Instead of one IV, there can be several IV and all contibute to some extent , to explain the variation in DV. Main focus of MLR :- 
####   (i) To study the relationship between IVs and DV
####  (ii) Which IV contributes the most to the explaination in variation of DV
#### (iii) Prediction of DV

## Adjusted R squared $(R_a^2)$ 

#### Including more predictors in a MLR model can artificially inflate R^2. Therefore adj. R^2 accounts for the the number of predictors in the model, it's a relative measure of fit, and should always be used in case of MLR instead of simple R^2

$$adj R^2 = 1 - \frac{\hat{\sigma}^2}{s^2_Y} $$

$ \hat{\sigma}^2 = $ estimated error variance $ = \frac{SSE}{n-p-1} $

This is also the MSE (mean squared error)

$s^2_Y = $ sample estimate of the variance of Y $ = \frac{SST}{n-1} $

$\tilde{\sigma}^2 =$ Maximum Likelihood Estimate of variance of error $= \frac{SSE}{n}$

$\tilde{\sigma}^2 < \hat{\sigma}^2$

#### VIF - Variance Inflation Factor is a way to measure the multicollinearity. It assesses how much the variance of an estimated regressor coefficient increases if the IVs are correlated. If no factors are correlated ,VIF will all be 1. 10>VIF>5 indicates high correlation that may be problematic. VIF > 10 indicates that the regressor coefficients are porrly estimated due to multicollinearity.

#### Mallows Cp must be low and approximately equal to the no. of parameters in the regression model. Models with lack-of-fit and bias have values of Mallows' Cp larger than p.

## CUSTOM CODE FOR MLR

```python
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values

# OneHot Encoding nominal variable State
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(drop='first'), [3])], remainder='passthrough') # without dummy variable
X = ct.fit_transform(X)
```

```python
X[:10]
```

```python
ct.get_feature_names()
```

```python
# splitting dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, shuffle=False)
```

```python
D1, D2, X1, X2, X3 = X_train[:,0], X_train[:, 1], X_train[:, 2], X_train[:, 3], X_train[:, 4] # seperating each IV for 
                                                                                            #ease of use in below code
```

```python
#model 

alpha = 0.0001
epochs = 2000
b0, b1, b2, b3, b4, b5 = 0,0,0,0,0,0


while epochs>0:
    
    y_pred = b0 + b1*D1 + b2*D2 + b3*X1 + b4*X2 + b5*X3
    error = y_pred - Y_train
    sse = sum(error**2)
    
    b0 = b0 - (alpha * 2 * sum(error))
    b1 = b1 - (alpha * 2 * sum(error * D1))
    b2 = b2 - (alpha * 2 * sum(error * D2))
    b3 = b3 - (alpha * 2 * sum(error * X1))
    b4 = b4 - (alpha * 2 * sum(error * X2))
    b5 = b5 - (alpha * 2 * sum(error * X3))
    
    if epochs%5==0:
        print(f"epoch : {2000-epochs}, SSE : {sse}")
        
    epochs -= 1
```

```python
# using matrices

alpha = 0.0001
epochs = 2000
B = np.ones(6)             # coefficients/parameters
H = np.insert(X_train.transpose(), 0, np.ones(40), 0)     # features


while epochs>0:
    
    y_pred = B.dot(H)
    error = y_pred - Y_train
    sse = sum(error**2)
    gradient = 2 * H.dot(error)
    B = B - alpha*gradient
    
    if epochs%5==0:
        print(f"epoch : {2000-epochs}, SSE : {sse}")   
        
    epochs -= 1
```
