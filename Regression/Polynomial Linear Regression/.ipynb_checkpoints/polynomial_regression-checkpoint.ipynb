## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import Image,YouTubeVideo
```

## Importing the dataset

```python
df = pd.read_csv('Position_Salaries.csv')
df
```

```python
X = df.iloc[:, 1].values.reshape(-1,1)
Y = df.iloc[:, -1].values
```

### Polynomial Linear Regression Pipeline

```python
Image('CAPTURE2.PNG', width=700,height=450, embed=True)
```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
```

```python
lin_reg.coef_, lin_reg.intercept_
```

$$\hat{Y} = \beta_0 + \beta_1Pos$$

$$\hat{Y} = -195333.33 + 80878.787Pos$$

## Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures

#feature transformation
poly_tranformer = PolynomialFeatures(degree=4, include_bias=False)
X_poly = poly_tranformer.fit_transform(X)

#model training
poly_lin_reg = LinearRegression()
poly_lin_reg.fit(X_poly, Y)
```

```python
# If there were more than 1 feature, say n features in X, PolynomialFeatures, for n predictors, X_poly would include all 
# the possible combinations of different order polynomials.This is known as Multi-dimensional Polynomial Regression.
# For eg, ig n=2 and PolynomialFeatures(degree=3), then :-
```

$$\hat{Y} = \beta_0 + \beta_1x_1 + \beta_2x_2 + \beta_3x_1x_2 + \beta_4x_1x_2^2 + \beta_5x_1^2x_2 + \beta_6x_1^3 + \beta_7x_2^3$$

```python
# But, there is a major issue with multi-dimensional Polynomial Regression – multicollinearity. Multicollinearity is the 
#interdependence between the predictors in a multi-dimensional regression problem. This restricts the model from fitting 
# properly on the dataset.
```

### multicollinearity ....??!! Doesn't multicollinearity affect the polynomial regression?? 
<font size='3' color='green'> multicollinearity (also collinearity) is a phenomenon in which one predictor variable in a multiple regression model can be <b><u>linearly</u></b> predicted from the others with a substantial degree of accuracy. 

Of course, Pos^1,Pos^2,Pos^3 & Pos^4 are correlated. However, one is not a linear transformation of the other by definition. Therefore, multicollinearity is not a problem here.</font>

```python
 poly_lin_reg.intercept_, poly_lin_reg.coef_
```

$$\hat{Y} = \beta_0 + \beta_1Pos +\beta_2Pos^2 + \beta_3Pos^3 + + \beta_4Pos^4$$

$$\hat{Y} = 184166.66 + (-211002.33)Pos + 94765.44Pos^2 + -15463.28Pos^3 + 890.15Pos^4$$

<font size='3'><b>The reason its a Linear model even though the variables/features are polynomials is because, in regards to the model & the coefficients, it is still linear. The $\beta$'s are not polynomials. The dataset however would be considered as non-linear relationship b/w X's & Y. The model just expresses the non-linear relationship b/w X's & Y using a linear model (the $\beta$'s). So when talking about the dataset, that's definetly non-linear but when we're talking about the class of regression, we're essentially talking about the unknown parameter's which in this case are linear.</b> </font>

### Example of a non-linear model:
<font size='4'>$$ \hat{Y} = \beta_0 X_1 + \frac{\beta_1}{(\beta_2X3 + \beta_3X4)}X2 + \epsilon  $$ </font>

### OR
<font size='4'> $$ \hat{Y} = \beta_0 + e^{\beta_1X1} + \log_{10} \beta_2X2 $$ </font>

## Visualising the Linear Regression results

```python
lin_reg.predict(X)
```

```python
%matplotlib notebook
plt.scatter(X, Y, color='red')
plt.plot(X, lin_reg.predict(X), color='green')
plt.plot(X)
plt.grid()
plt.title('Truth or BLuff (SLR)')
plt.xlabel('Position', fontsize=13)
plt.ylabel('Salary', fontsize=13)
plt.show()
```

### $R^2$

```python
from sklearn.metrics import r2_score
r2_score(Y, lin_reg.predict(X))
```

## Visualising the Polynomial Regression results

```python
poly_lin_reg.predict(X_poly)
```

```python
%matplotlib notebook
plt.scatter(X, Y)
plt.plot(X, poly_lin_reg.predict(X_poly), color='green')
plt.grid()
plt.title('Truth or Bluff (PLR)')
plt.xlabel('Postion', fontsize=13)
plt.ylabel('Salary', fontsize=13)
plt.show()
```

### $R^2$

```python
#with intercept term
r2_score(Y, poly_lin_reg.predict(X_poly))
```

```python
#without intercept term i.e fit_intercept=False, accuracy is slightly less as even all the predictions are sligtly less that
# the model with fit_intercept=True
r2_score(Y, poly_lin_reg.predict(X_poly))
```

```python
plt.scatter(X, Y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue', label='SLR')
plt.plot(X, poly_lin_reg.predict(X_poly), color='green', label='PLR')
plt.grid()
plt.title('Truth or BLuff (SLR vs PLR)')
plt.xlabel('Position', fontsize=13)
plt.ylabel('Salary', fontsize=13)
plt.legend()
plt.show()
```

## How to select the correct order

### Bayes Information Criteria (BIC), thanks to Bayes

$$BIC_k = n\log{(SSE)} + k\log{(n)}$$

$n = no. of datapoints$

$k = no. of parameters/betas$

#### $BIC_k$ is calculted for every model with different orders. Then a a graph is plotted of Polynomial model order v/s BIC. Looking 
#### at the plot, minimum BIC level is found and that model with corresponding polynomial order is selected. (Beware of overfitting though)

## Q) How to know if the dataset is linear or non linear & then decide whether it would be most appropriate to use a Polynomial Linear Regression model or not if there are more than 3 features (Since we can't visualize higher dimensional data).

<font size='5' color='green'><b> 2 Simple methods:</b></font>

<font size='4' color='green'>(i) First, fit a linear model to the data and then check the scedasticity => If homoscedastic, then linear data, if heteroscedastic, then non-linear model would fit the data well. Similar & more confirming approach could be to set a <b>threshold</b> for error variance, if error variance of the linear model > <b>threshold</b>, non-linear model to the rescue. 
    
(ii) You can check the p-value for the coefficients of the exponential terms (Pos^2,...,Pos^n). If they're significant, it further confirms the non-linear relationship b/w the original features & target variable. </font>
